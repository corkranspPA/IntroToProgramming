# Lab 2 - StringHelper

## Task

Create a javascript file named `stringHelper.js` with the following 8 functions:

### endLy

Given a string, return true if it ends in 'ly', and false otherwise.

- endLy('oddly') → true
- endLy('y') → false
- endLy('oddy') → false

### proctorColor

Given a string of any length, if the string begins with 'white' or 'green' return that color string, otherwise return the empty string.

- proctorColor('greenxx') → 'green'
- proctorColor('xxwhite') → ''
- proctorColor('whiteTimes') → 'white'

### missingChar

Given a non-empty string and a Number variable `n`, return a new string where
the character at index `n` has been removed. The value of `n` will be a valid
index of a character in the original string (i.e. `n` will be in the range
0..str.length-1 (inclusive)).

- missingChar('kitten', 1) → 'ktten'
- missingChar('kitten', 0) → 'itten'
- missingChar('kitten', 4) → 'kittn'

### lastFirst

Given a string, return a new string where the first and last characters have been
exchanged.  The given string will have a length of at least 2.

- lastFirst('code') → 'eodc'
- lastFirst('abc') → 'cba'
- lastFirst('ab') → 'ba'

### fizzString

Given a string `str`, if the string starts with 'f' return 'Fizz'. If the string ends with 'b' return 'Buzz'. If both the 'f' and 'b' conditions are true, return 'FizzBuzz'. In all other cases, return the string unchanged.

- fizzString('fig') → 'Fizz'
- fizzString('dib') → 'Buzz'
- fizzString('fib') → 'FizzBuzz'
- fizzString('Proctor') → 'Proctor'

### isDoubled

Given a string of even length, return whether the first half is the same as the
last half.

- isDoubled('BooBoo') → true
- isDoubled('HelloThere') → false
- isDoubled('aa') → true
  
### middleThree

Given a string of odd length, return the string length 3 from its middle, so
'Candy' yields 'and'. The string length will be at least 3.

- middleThree('Candy') → 'and'
- middleThree('and') → 'and'
- middleThree('solving') → 'lvi'
