# Lab 2 - StringHelper Redux

## Task

Create a javascript file named `stringHelper.js` with the following functions:

### firstLetter

Given a string with a length of at least 1, return the first letter of the string.

### thirdLetter

Given a string with a length of at least 3, return the third letter of the string.

### thirdLetterS

Given a string with a length of at least 3, return true if the third letter is
's', and false otherwise.

### lastLetter

Given a string with a length of at least 1, return the last letter of the
string.

### lastLetterS

Given a string with a length of at least 1, return true if the last letter of
the string is 's', and false otherwise.

### startsGood

Given a string with a length of at least 4, return true if the string starts
with 'good', and false otherwise.

### startsGoodBad
Given a string, return true if the string starts with 'good' or 'bad', and false otherwise.

### almostStartsGood

Given a string with a length of at least 4, return true if the string contains
the string 'good' starting at index 1, and false otherwise.

### hasBad

Given a string, return true if 'bad' appears starting at index 0 or 1 in the
string (e.g. 'badxxx' or 'xbadxx', but NOT 'xxbadxx'!), and false otherwise.

### combineEnds

Given 2 strings, `a` and `b`, return a new string made of the first char of a and the last char of b, so "yo" and "java" yields "ya".

### nSlice

Given a string and a Number `n`, return a string made of the first and last `n` chars from the string. The string length will be at least `n`.

- nSlice("Hello", 2) → "Helo"
- nSlice("Chocolate", 3) → "Choate"
- nSlice("Chocolate", 1) → "Ce"
