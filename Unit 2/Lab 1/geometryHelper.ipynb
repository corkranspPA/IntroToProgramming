# Lab 1 - GeometryHelper

## Task

We will be practicing creating functions and using the built-in functions in the
Math object in order to create simple programs to assist with geometry work.

Create a javascript file named `geometryHelper.js` with the following functions:

### areaOfRectangle

A familiar function, to warm up with!  Given the `length` and `width` of a rectangle, `return` the
area.  The formula is `length * width`.

- areaOfRectangle(1,1) → 2
- areaOfRectangle(4,6) → 24
- areaOfRectangle(0,99) → 0

### perimeterOfRectangle

Given the `length` and `width` of a rectangle, `return` the
perimeter.  The formula is `(length * 2) + (width * 2)`.

- perimeterOfRectangle(1,1) → 4
- perimeterOfRectangle(4,6) → 20
- perimeterOfRectangle(0,99) → 198

### areaOfCircle

Given the `radius` of a circle, `return` the area of the circle.  The formula is
`pi * radius * radius`.  Make sure to use `Math.PI` in your code.

- areaOfCircle(1) → 3.141592653589793
- areaOfCircle(2.5) → 19.634954084936208
- areaOfCircle(0) → 0

### circumferenceOfCircle

Given the `radius` of a circle, `return` the circumference of the circle.  The formula is
`2 * pi * radius`.  Make sure to use `Math.PI` in your code.

- circumferenceOfCircle(1) → 6.283185307179586
- circumferenceOfCircle(2.5) → 15.707963267948966
- circumferenceOfCircle(0) → 0

### volumeOfSphere

Given a radius, `return` the volume of the sphere.  The formula is
`(4 / 3) * pi * radius * radius * radius`.  Please use `Math.PI` in your
code, as well as using `Math.pow` to raise radius to the third power ('radius cubed').

- volumeOfSphere(1) → 4.1887902047863905
- volumeOfSphere(2.5) → 65.44984694978736
- volumeOfSphere(0) → 0

### perimeterOfTriangle

Given a 3 side lengths of a triangle, `a`, `b`, and `c`, `return` the perimeter of
the triangle.  The formula is `a + b + c`.  This is really just a helper
function for the last function you will create.

- perimeterOfTriangle(1,1,1) → 3
- perimeterOfTriangle(5,6,7) → 18
- perimeterOfTriangle(0,0,0) → 0

### areaOfTriangle

Given a 3 side lengths of a triangle, `a`, `b`, and `c`, `return` the area of
the triangle.  The formula ("Heron's formula") is `√s(s - a)(s - b)(s - c)`,
where `s` is **half the perimeter**, or `(a + b + c)/2`.  `Math.sqrt` can be
used to find the square root.  The best solutions will use the `perimeterOfTriangle` method you've already created.

- areaOfTriangle(1,1,1) → 0.4330127018922193
- areaOfTriangle(5,6,7) → 14.696938456699069
- areaOfTriangle(0,0,0) → 0
